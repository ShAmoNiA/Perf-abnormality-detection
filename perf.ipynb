```python
import os
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
directory = '/content/drive/MyDrive/Data_mainProj'
MainData = os.listdir(directory)
```

```python
def pandaGen(pidPath,pids):
  sysCall = {}
  for pid in pids:
    file = pidPath+"/"+pid+"/perf.data.txt"
    f = open(file, "r")
    lines = f.readlines()
    for line in lines:
      split = line.split(":")
      for counter_ in range(len(split)):
          if("syscalls" in split[counter_]):
            sysName = split[counter_+1]
            if "enter" in sysName:
              key = split[0].replace(" ","").split("[")[0]
              key_ = ''.join(i for i in key if not i.isdigit())
              if key_ in sysCall:
                sysCall[key_].append(sysName)
              else:
                sysCall[key_] = [sysName]

  df = []
  for key in sysCall:
    unique, counts = np.unique(np.array(sysCall[key]), return_counts=True)
    hold = dict(zip(unique, counts))
    d = hold.copy()
    d.update({"name":key})
    df_hold = pd.DataFrame.from_dict([d])
    df_hold = df_hold[ ['name'] + [ col for col in df_hold.columns if col != 'name' ] ]
    df.append(df_hold)
  output = pd.DataFrame()
  for hold in df:
      output = pd.concat([output, hold], ignore_index=True)

  return output
```

```python
task_out =[]
for task in MainData:
  print(task)
  records = os.listdir(directory+"/"+task)
  output_rec = []
  for record in records:
    pidPath = directory+"/"+task+"/"+record+"/syscalls"
    try:
      pids = os.listdir(pidPath)  
      output_rec.append(pandaGen(pidPath,pids))
    except:
      print("error in "+pidPath)
  task_out.append(output_rec)
```

```python
columns_name = []
for task in task_out:
  for rec in task:
    for item in rec.columns:
      if item not in columns_name:
        columns_name.append(item)
```

```python
proc_name = []

for task_counter in range(len(task_out)):
  for rec_counter in range(len(task_out[task_counter])):
    for proc in task_out[task_counter][rec_counter]["name"]:
      if proc not in proc_name:
        proc_name.append(proc)
```

```python
for task_counter in range(len(task_out)):
  for rec_counter in range(len(task_out[task_counter])):
    holdProc = []
    for proc in proc_name:
      if proc not in task_out[task_counter][rec_counter]["name"].tolist():
        holdProc.append(proc)
    df = pd.DataFrame(holdProc,columns =['name'])
    task_out[task_counter][rec_counter] = task_out[task_counter][rec_counter].append(df, ignore_index = True)
    
```

```python
df_ = pd.DataFrame(columns=columns_name)

for task_counter in range(len(task_out)):
  for rec_counter in range(len(task_out[task_counter])):
    task_out[task_counter][rec_counter] = pd.concat([df_,task_out[task_counter][rec_counter]], ignore_index=True).sort_values(by=['name']).reset_index(drop=True).fillna(0)
```

```python
task_out_list = []
main = []
for task_counter in range(len(task_out)):
  rec_out_list=[]
  for rec_counter in range(len(task_out[task_counter])):
    hold = task_out[task_counter][rec_counter].drop('name', axis=1).to_numpy().reshape(-1)
    rec_out_list.append(hold)
    main.append(hold)
  task_out_list.append(rec_out_list)
```
